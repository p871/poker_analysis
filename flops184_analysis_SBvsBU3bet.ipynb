```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('SBvsBU3bet.csv', index_col=0)
```

```python
# borad typeによって分割
dfr = df[df['board_type'] == 'rainbow']
dft = df[df['board_type'] == 'twotone']
dfm = df[df['board_type'] == 'monotone']
```

```python
dfr.groupby('high_card').mean().sort_values('check',ascending=False)
```

```python
# board typeに関係なく並べた場合のcheck率
order_list = [*'456789TJQKA']
sns.barplot(x='high_card', y='check', data=df, ci=None, order=order_list)
```

```python
# rainbow boardのcheck率。high cardがT以上の場合はほとんどcheckしない。
sns.barplot(x='high_card', y='check', data=dfr, ci=None, order=order_list)
```

```python
# rainbow boardの66% sizeのbet率。意外にもhigh cardがAの時に最も使わない。
sns.barplot(x='high_card', y='bet66%', data=dfr, ci=None, order=order_list)
```

```python
# rainbow boardの33% sizeのbet率。high cardがQ以上の場合高頻度で行われる。
sns.barplot(x='high_card', y='bet33%', data=dfr, ci=None, order=order_list)
```
